```python
# "parse_dates" converte o campo para o tipo data
# "dayfirst=True" Em uma data indica no campo de data que o primeiro bloco é de dia 

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Digital Innovation One/13 - Fundamentos de ETL com Python/ocorrencia_2010_2020.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
```

**Cria um Schema para o Dataframe**

```python
schema = pa.DataFrameSchema(
    columns = {
        "ColunaInexistente": pa.Column(pa.Int, required=False), ## Caso a coluna não exista no arquivo ela se torna opcional, assim passa na validação do Schema
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2)), # Verifica se tem no minimo 2 ou no maximo 2
        "ocorrencia_aerodromo": pa.Column(pa.String),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)   
    }
)
```

**Faz a validação do schema criado. Retorna o dataframe caso o tipo de dados especificaado seja valido**

```python
schema.validate(df)
```
